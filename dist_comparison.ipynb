```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import os
import cv2

def plot_squares_with_numbers(arr):
    viridis = matplotlib.colormaps['viridis']
    color_div, color_add = np.nanmax(arr) - np.nanmin(arr), np.nanmin(arr)
    dim1, dim2 = arr.shape[0], arr.shape[1]
    fig, ax = plt.subplots(figsize=(32, 32))
    for i in range(dim1):
        for j in range(dim2):
            color = viridis((arr[i][j] - color_add) / color_div)
            ax.add_patch(plt.Rectangle((j, dim1 - i - 1), 1, 1, color=color))
            ax.text(j + 0.5, dim1 - i - 1 + 0.5, str(round(arr[i][j], 2)), ha='center', va='center')

    ax.set_xlim(0, dim2)
    ax.set_ylim(0, dim1)
    ax.axis('off')
    plt.show()
```

```python
data_path = "gogs_test_data_real/day_2_obj/depth"
real_img = np.loadtxt(os.path.join(data_path, "1.txt"))
plt.imshow(real_img)
plt.show()
real_img = cv2.resize(real_img[200:450, 1000:1200], (32,32))
print(real_img.shape)
plot_squares_with_numbers(real_img)
```

```python
dex3_path = "dataset/dexnet_3/dexnet_09_13_17/tensors/depth_ims_tf_table_00300.npz"
dex3_depth = np.load(dex3_path)["arr_0"][100].squeeze()
print(dex3_depth.shape)
plot_squares_with_numbers(dex3_depth)
```
